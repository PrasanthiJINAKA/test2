# Plotting for exploratory data analysis (EDA)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#load Iris.csv into a pandas data frame
irisFlwrs = pd.read_csv("iris.csv")
```

```python
#how many data-points and features
print(irisFlwrs.shape)
```

```python
#what are the column names in our dataset
print (irisFlwrs.columns)
```

```python
#how many data points  for each class are present? (or) How many flowers for each species are present
irisFlwrs["species"].value_counts()

#balanced dataset vs imbalanced datasets
#IrisFlwrs is balanced dataset
```

# 2-D Scatter plot

```python
#2-D Scatter plot
#Always understand the axis : labels and scale

irisFlwrs.plot(kind='scatter', x='sepal_length', y='sepal_width');
plt.show()
```

```python
#Seaborn
sns.set_style("whitegrid")
sns.FacetGrid(irisFlwrs, hue="species", size=4) \
   .map(plt.scatter, "sepal_length", "sepal_width") \
   .add_legend();
plt.show();
```

**Observation(s):**
1. Using sepal_length and sepal_width features, we can distinguish Setosa flowers from others.
2. Seperating Versicolor from Viginica is much harder as they have considerable overlap.

```python
#pairwise scatter plot: Pair-Plot
# Dis-advantages: 
#Can be used when number of features are high.
#Cannot visualize higher dimensional patterns in 3-D and 4-D. 

plt.close()
sns.set_style("whitegrid")
sns.pairplot(irisFlwrs, hue="species", height=3);
plt.show()
```
